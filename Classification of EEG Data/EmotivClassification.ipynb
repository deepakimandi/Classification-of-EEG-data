```python
y = df.Target
X = df.drop(['Target'], axis = 'columns')
```

```python
# C = 1200, gamma = 0.00033
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

# modelSVM = SVC(C = 1300, gamma = 0.00033)
# modelSVM.fit(X_train, y_train)

modelRFC = RandomForestClassifier(n_estimators = 60)
modelRFC.fit(X_train, y_train)
```

```python
print('Accuracy : ' + str(modelRFC.score(X_test, y_test)))

from statistics import mean
from sklearn.model_selection import cross_val_score
result = cross_val_score(modelRFC, X, y)
print('Accuracies on applying 5-fold cross-validation : ' + str(result))
print('Mean accuracy : ' + str(mean(result)))
```

```python
y_predicted = modelRFC.predict(X_test)
print('Number of test samples : ' + str(len(X_test)))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
conmat = confusion_matrix(y_test, y_predicted)
print(conmat)
```
